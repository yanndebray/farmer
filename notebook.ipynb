# The Farmer Was Replaced (by ChatGPT)

```python
import openai, os, base64, io
from PIL import ImageGrab
from dotenv import load_dotenv
load_dotenv()
```

```python
with open('inst.txt', 'r') as file:
    inst = file.read()
```

```python
def clip(name):
    ImageGrab.grabclipboard().save(name+'.png', 'PNG')
```

```python
def chat(prompt):
    res = openai.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": inst},
            {"role": "user", "content": prompt},
        ]
    )
    return res.choices[0].message.content
```

```python
def chat_vision(prompt, base64_image):
    response = openai.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": inst},
            {
            "role": "user",
            "content": [
                {"type": "text", "text": prompt},
                {"type": "image_url","image_url": {"url": f"data:image/jpeg;base64,{base64_image}"}},
            ],
            }
        ]
    )
    return response.choices[0].message.content
```

```python
def encode_image(image):
    buffered = io.BytesIO()
    image.save(buffered, format="PNG")
    return base64.b64encode(buffered.getvalue()).decode('utf-8')
```

```python
def save_chat(prompt, res):
    with open(f'chats/{prompt}.txt', 'w') as file:
        file.write(res)
```

```python
prompt = "how to plant pumpkins"
res = chat(prompt)
print(res)
```

```python
save_chat(prompt, res)
```

```python
# get image from clipboard
img = ImageGrab.grabclipboard()
img
```

```python
prompt = "How to navigate the maze"
res = chat_vision(prompt,encode_image(img))
print(res)
```
